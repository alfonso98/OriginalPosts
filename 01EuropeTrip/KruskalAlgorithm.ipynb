```
def isInArray(arr, x):
    resp = False
    for z in arr:
        if( type(z) == str ):
            continue
        if( z == x ):
            resp = True
        
    return(resp)
```

```
def minimumSpanningTree(data):
    # Initial values
    start = 0
    n = len(data)
    route = [start]
    nodes = []
    prices = []
    cost = 0

    for i in range(0, n):
        if i != start:
            nodes.append(i)


    # Start Kruskal Algorithm

    while len(nodes) > 0:
        aux = []
        for i in route:
            for j in range(0,n):
                if type(data[i][j]) != str :
                    if isInArray(route, j) == False:
                        aux.append((i, j, data[i][j]))

        small = aux[0]

        for elem in aux:
            if elem[2] <= small[2]:
                small = elem

        cost += small[2]
        prices.append(small)
        route.append(small[1])
        nodes.remove(small[1])

    print('General cost of the tree: ', cost)
    print('\nRoutes: ')
    for elem in prices:
        print('Node ', elem[0]+1, ' with node ', elem[1]+1, ' with a cost of ', elem[2])
```
